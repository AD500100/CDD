```python
from re import X
# total no of member from vc
# Convert 'MEMBER' column to numeric, handling non-numeric values as errors
df['MEMBER'] = pd.to_numeric(df['MEMBER'], errors='coerce')

# Group by 'VILLAGE CO-ORDINATOR' and sum the numeric 'MEMBER' values
df = df.groupby(['VILLAGE CO-ORDINATOR'], as_index=False)['MEMBER'].sum().sort_values(by='MEMBER', ascending=False)

# Create the bar plot first
ax = sns.barplot(data = df , x= 'VILLAGE CO-ORDINATOR', y= 'MEMBER', color= 'green', hue='VILLAGE CO-ORDINATOR' )
plt.title('Total Member')
plt.xticks(rotation=90) # Rotate x-ticks after creating the plot

# Now you can iterate over the patches
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'),
                   (p.get_x() + p.get_width() / 2., p.get_height()+5),
                   ha = 'center', va = 'center')
ax.set_facecolor('yellow')
sns.set(rc={'figure.figsize':(15, 8)})
plt.show() # Display the plot
```

```python
path = '/content/drive/MyDrive/Colab CDD DATA - MASTER DASHBOARD - CDD DATA - MASTER DASHBOARD.csv'
df = pd.read_csv(path)
df.head()
```

```python
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the data
# Replace with actual column names if they are different
# Assuming you want to sum the 'POULTRY' column

UPGRADE_MICRO = df['UPGRADE MICRO'].sum()
NEW_MICRO = df['NEW MICRO'].sum()
SKILL_DEVELOPMENT = df['SKILL DEVELOPMENT'].sum()
#SHG_TRAININGS = df['SHG TRAININGS'].sum()
POULTRY = df['POULTRY'].sum()

# Creating the pie chart
labels = ['UPGRADE MICRO', 'NEW MICRO', 'SKILL DEVELOPMENT', 'POULTRY']
sizes = [UPGRADE_MICRO, NEW_MICRO, SKILL_DEVELOPMENT, POULTRY]
colors = ['#ff9999','#66b3ff','#99ff99', '#c2c2f0','#ffb3e6']  # Added more colors for the new slices

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

# Adding a circle at the center to make it a donut chart
centre_circle = plt.Circle((0,0),0.25,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
fig.patch.set_facecolor('yellow')
ax1.axis('equal')
plt.tight_layout()
plt.title('Distribution of Activities')
plt.show()
```
